```python
# Boxplot to identify outliers across all features in the Parkinson's dataset
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("Outlier Detection Across Dataset Features")
plt.xlabel("Features")
plt.ylabel("Value Range")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Identify and remove outliers using the IQR method
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

# Define extended IQR bounds
lower_bound = q1 - 1.8 * iqr
upper_bound = q3 + 1.8 * iqr

# Filter out rows with any value outside the bounds
parkinsons_df_no_outliers = df[
    ~((df < lower_bound) | (df > upper_bound)).any(axis=1)
]

# Report how many rows were removed
print(f"Original dataset size       : {df.shape[0]}")
print(f"Dataset size after cleaning : {parkinsons_df_no_outliers.shape[0]}")
print(f"Total outliers removed      : {df.shape[0] - parkinsons_df_no_outliers.shape[0]}")
```

```python
# Normalize pitch-related features using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Apply scaling to selected frequency features
frequency_features = ["MDVP:Fo(Hz)", "MDVP:Fhi(Hz)", "MDVP:Flo(Hz)"]
df[frequency_features] = scaler.fit_transform(df[frequency_features])

# View the updated DataFrame
df.head()
```

# Feature Scaling 

```python
x = df.drop(columns=['status'])
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
print(x_train)
```

# logistic regression

```python
# Train and evaluate a Logistic Regression model for Parkinson's disease classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Initialize and train the model
model = LogisticRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate performance
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Display the metrics
print(f"Precision Score     : {precision:.4f}")
print(f"Recall Score        : {recall:.4f}")
print(f"F1 Score            : {f1:.4f}")
print(f"ROC AUC Score       : {auc:.4f}")
```

# Decision Tree Classifier

```python
# Train and evaluate a Decision Tree classifier for Parkinson's disease prediction
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Initialize the model with a maximum depth of 2
model = DecisionTreeClassifier(max_depth=2)
model.fit(x_train, y_train)

# Generate predictions
y_pred = model.predict(x_test)

# Calculate evaluation metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Display the metrics
print(f"Precision Score     : {precision:.4f}")
print(f"Recall Score        : {recall:.4f}")
print(f"F1 Score            : {f1:.4f}")
print(f"ROC AUC Score       : {auc:.4f}")
```

# Random Forest classification

```python
# Train and evaluate a Random Forest classifier for Parkinson's disease prediction
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Calculate performance metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Display evaluation results
print(f"Precision Score     : {precision:.4f}")
print(f"Recall Score        : {recall:.4f}")
print(f"F1 Score            : {f1:.4f}")
print(f"ROC AUC Score       : {auc:.4f}")
```

# K Nearest Neignbors classification

```python
# Train and evaluate a K-Nearest Neighbors classifier for Parkinson's disease prediction
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Initialize and fit the model
model = KNeighborsClassifier()
model.fit(x_train, y_train)

# Generate predictions
y_pred = model.predict(x_test)

# Calculate performance metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Display the results
print(f"Precision Score     : {precision:.4f}")
print(f"Recall Score        : {recall:.4f}")
print(f"F1 Score            : {f1:.4f}")
print(f"ROC AUC Score       : {auc:.4f}")
```

# Support Vector Classifier

```python
# Train and evaluate a Support Vector Machine (SVM) model for Parkinson's disease prediction
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Initialize and fit the SVM model
model = SVC(probability=True, kernel='rbf', random_state=42)
model.fit(x_train, y_train)

# Generate predictions
y_pred = model.predict(x_test)

# Compute evaluation metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Display the metrics
print(f"Precision Score     : {precision:.4f}")
print(f"Recall Score        : {recall:.4f}")
print(f"F1 Score            : {f1:.4f}")
print(f"ROC AUC Score       : {auc:.4f}")
```
